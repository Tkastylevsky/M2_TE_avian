```python
import pandas as pa
import itertools
import csv
names_col = ['SW_score',"perc_div","perc_del","perc_ins", "query_seq", "pos_query_begin", "pos_query_end", "pos_query_left","strand", "matching_repeat", "repeat_class_family", "pos_in_repeat_begin", "pos_in_repeat_end", "pos_in_repeat_left", "ID","star"]
```

```python
data = pa.read_csv("/home/tkastylevsky/FASTA_files/Gallus_gallus.GRCg6a.dna.chromosome.16.fa.out",header = None, skiprows = 2, sep = '\s+', names = names_col)
data['strand'] = data['strand'].replace('C','-')
ref = pa.read_csv("/home/tkastylevsky/FASTA_files/chkchr16_refgalGal6", sep = '\s+')
ref = ref.drop(["#bin","repFamily"], axis = 1)
ref.insert(15,"star", 'NaN')
ref.columns = names_col
ref.insert(16,"Type", 'Ref')
data.insert(16, "Type", 'Data')
```

```python
analysis_df = data.append(ref)
analysis_df.insert(17,"ID_2",list(range(0, len(analysis_df))))
```

```python
analysis_df['intersect'] = analysis_df.apply(lambda row: len(analysis_df[ ( ( (row.pos_query_begin <= analysis_df.pos_query_begin) & (analysis_df.pos_query_begin <= row.pos_query_end) ) \
                                            | ( (analysis_df.pos_query_begin <= row.pos_query_begin) & (row.pos_query_begin <= analysis_df.pos_query_end) ) )
                           & (row.ID_2 != analysis_df.ID_2) & (row.Type != analysis_df.Type) & (row.repeat_class_family == analysis_df.repeat_class_family) ]),axis=1)
```
