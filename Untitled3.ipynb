```python
import pandas as pa
import numpy as np
import itertools
import csv
```

```python
def out_to_gff(path,symbol):
    names_col_data = ['SW_score',"perc_div","perc_del","perc_ins", "query_seq", "pos_query_begin", "pos_query_end", "pos_query_left","strand", "matching_repeat", "repeat_class_family", "pos_in_repeat_begin", "pos_in_repeat_end", "pos_in_repeat_left", "TE_"+symbol+"_ID","star"]
    data = pa.read_csv(path,header = None, skiprows = 2, sep = '\s+', names = names_col_data)
    data['strand'] = data['strand'].replace('C','-')
    ID = symbol+"_ID"
    data.insert(1,ID,list(range(1,len(data)+1)))
    data_split = data.repeat_class_family.str.split('/',expand = True)
    data = data.join(data_split)
    data.columns = ['SW_score',ID,"perc_div","perc_del","perc_ins", "query_seq", "pos_query_begin", "pos_query_end", "pos_query_left","strand", "matching_repeat", "repeat_class_family", "pos_in_repeat_begin", "pos_in_repeat_end", "pos_in_repeat_left", "TE_"+symbol+"_ID","star","repeat_class"+symbol,"repeat_family"+symbol]
    data = data.drop('repeat_class_family', axis = 1)
    return data
```

```python
path = "/home/tkastylevsky/FASTA_files/repeatmasker/gallus_gallus_proper/gallus_chr1_lib_whole_genome/chr1.fa.out"
symbol = "A"
out_to_gff(path,symbol)
```
